```python
import numpy as np
from matplotlib import pyplot as plt
from NearFieldOptics import Materials as M
from NearFieldOptics import TipModels as T
```

```python
freq = 1500
qs = np.linspace(0, 20, 500)
a = 30E-7

SLG = M.SingleLayerGraphene(chemical_potential=10, gamma=100)
layers = M.LayeredMedia(SLG, (M.BN_Caldwell, 43E-7), exit=M.SiO2_300nm)
layers_TM = M.LayeredMediaTM(SLG, (M.BN_Caldwell, 43E-7), exit=M.SiO2_300nm)

Rp = lambda qnorm: layers.reflection_p(freq,qnorm*1/a)
Rp_TM = lambda qnorm: layers_TM.reflection_p(freq,qnorm*1/a)

rps = Rp(qs)
rps_TM = Rp_TM(qs)

plt.figure()
plt.plot(qs, np.imag(rps), label = 'LayeredMedia')
plt.plot(qs, np.imag(rps_TM), label = 'LayeredMediaTM')
plt.xlabel('q / 1/a'); plt.ylabel('Im(r_p)')
plt.title('LayeredMedia vs LayeredMediaTM')
plt.legend()
```
